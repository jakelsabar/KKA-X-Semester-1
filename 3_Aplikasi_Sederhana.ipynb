#Aplikasi Sederhana Menghitung Volume Bola

#Aplikasi Sederhana Menghitung Volume Dan luas Permukaan Kecurut

```python
import math
# Volume: V = (1/3) * π * r² * h
# Luas permukaan: A = π * r * (r + s)

r = float(input("Masukkan jari-jari alas kerucut (r): "))
h = float(input("Masukkan tinggi kerucut (h): "))

volume = (1/3) * math.pi * r**2 * h
s = math.sqrt(r**2 + h**2)
luas_permukaan = math.pi * r * (r + s)


print(f"Volume kerucut: {volume:.2f}")
print(f"Luas permukaan kerucut: {luas_permukaan:.2f} cm^2. ")
```
