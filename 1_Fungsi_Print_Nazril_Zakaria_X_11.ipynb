```python
import datetime
from IPython.display import display, HTML

# List nama hari dan bulan dalam bahasa Indonesia
nama_hari = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]
nama_bulan = [
    "Januari", "Februari", "Maret", "April", "Mei", "Juni",
    "Juli", "Agustus", "September", "Oktober", "November", "Desember"
]

# Ambil tanggal sekarang (untuk hari, tanggal, bulan, tahun)
sekarang = datetime.datetime.now()
indeks_hari = sekarang.weekday()
hari = nama_hari[indeks_hari]
tanggal = sekarang.day
bulan = nama_bulan[sekarang.month - 1]
tahun = sekarang.year

# HTML dengan jam real-time WIB
html_code = f"""
<div style="
    font-family: Arial, sans-serif;
    background: linear-gradient(135deg, #74ebd5, #ACB6E5);
    padding: 20px;
    border-radius: 15px;
    width: 340px;
    text-align: center;
    box-shadow: 0px 4px 10px rgba(0,0,0,0.2);
">
    <h2 style="margin:0; color:#2C3E50;">ü§≠ Tanggal & Waktu (WIB) ü§≠</h2>
    <p style="font-size: 20px; margin: 5px 0; color:#34495E;">
        <b>{hari}</b>, {tanggal} {bulan} {tahun}
    </p>
    <p style="font-size: 26px; margin: 10px 0; color:#2C3E50;">
        ‚è∞ <span id="clock"></span> WIB ‚è∞
    </p>
</div>

<script>
function updateClock(){{
    var now = new Date();
    // Konversi ke WIB (UTC+7)
    var utc = now.getTime() + (now.getTimezoneOffset() * 60000);
    var wibTime = new Date(utc + (7 * 3600000));

    var jam = String(wibTime.getHours()).padStart(2,'0');
    var menit = String(wibTime.getMinutes()).padStart(2,'0');
    var detik = String(wibTime.getSeconds()).padStart(2,'0');

    document.getElementById("clock").textContent = jam + ":" + menit + ":" + detik;
}}
setInterval(updateClock, 1000);
updateClock();
</script>
"""

display(HTML(html_code))
```
